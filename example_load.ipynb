```python
import datacube
dc = datacube.Datacube()
```

```python
ds = dc.find_datasets(product="spot")
print(f"Found {len(ds)} datasets")
```

```python
# Change the dataset here to load the other one
data = dc.load(datasets=ds[0:1], output_crs="epsg:32735", resolution=(-10,10))
data
```

```python
# Can plot other bands or combine into an RGB image
data.nir.plot(robust=True)
```
